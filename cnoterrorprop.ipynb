```python
# X error propagation in bit basis
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# Create a 3-qubit circuit with no error (entangling the qubits)
qc_no_error = QuantumCircuit(3, 3) # Initialized in the |0> state
qc_no_error.cx(0, 1)  # CNOT gate (Q0 -> Q1)
qc_no_error.cx(1, 2)  # CNOT gate (Q1 -> Q2)
qc_no_error.measure([0, 1, 2], [0, 1, 2])  # Measure all qubits

# Create a 3-qubit circuit with an X error on Q1 (entangling the qubits)
qc_xbit_error = QuantumCircuit(3, 3) # Initialized in the |0> state
qc_xbit_error.x(1)  # Apply X error (bit-flip) to Q1
qc_xbit_error.cx(0, 1)  # CNOT gate (Q0 -> Q1)
qc_xbit_error.cx(1, 2)  # CNOT gate (Q1 -> Q2)
qc_xbit_error.measure([0, 1, 2], [0, 1, 2])  # Measure all qubits

# Use Aer simulator
simulator = Aer.get_backend('aer_simulator')

# Run the simulation for both circuits
result_no_error = simulator.run(qc_no_error, shots=10000).result()
result_with_error = simulator.run(qc_xbit_error, shots=10000).result()

# Get counts (measurement outcomes)
counts_no_error = result_no_error.get_counts()
counts_with_error = result_with_error.get_counts()

# Plot the histograms to visualize the impact of the X error
plot_histogram([counts_no_error, counts_with_error], legend=['No X Error', 'With X Error'],
               title= 'X Error in Bit Basis')

```

```python
# Create visualization of circuit
qc_xbit_error.draw('mpl')
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import numpy as np

# Function to add depolarizing noise to a CNOT gate manually
def add_depolarizing_noise(circuit, qubit1, qubit2, error_prob):
    # Apply depolarizing noise to the control qubit (qubit1)
    if np.random.rand() < error_prob:
        noise_gate = np.random.choice(['x', 'y', 'z'])
        getattr(circuit, noise_gate)(qubit1)
    # Apply depolarizing noise to the target qubit (qubit2)
    if np.random.rand() < error_prob:
        noise_gate = np.random.choice(['x', 'y', 'z'])
        getattr(circuit, noise_gate)(qubit2)

# Parameters
error_prob = 0.1  # Probability of depolarizing error
shots_per_run = 1000  # Number of shots per simulation
num_runs = 10  # Number of times to repeat the simulation

# Backend for simulation
backend = Aer.get_backend('qasm_simulator')

# Initialize cumulative counts for both scenarios
cumulative_counts_no_error = {}
cumulative_counts_with_error = {}

# Run the simulation in a loop
for _ in range(num_runs):
    # Circuit without depolarizing noise
    qc_no_error = QuantumCircuit(3, 3)
    qc_no_error.cx(0, 1)  # CNOT gate (Q0 -> Q1)
    qc_no_error.cx(1, 2)  # CNOT gate (Q1 -> Q2)
    qc_no_error.measure([0, 1, 2], [0, 1, 2])  # Measure all qubits
    
    # Circuit with depolarizing noise
    qc_error = QuantumCircuit(3, 3)
    qc_error.cx(0, 1)  # CNOT gate (Q0 -> Q1)
    add_depolarizing_noise(qc_error, 0, 1, error_prob)
    qc_error.cx(1, 2)  # CNOT gate (Q1 -> Q2)
    add_depolarizing_noise(qc_error, 1, 2, error_prob)
    qc_error.measure([0, 1, 2], [0, 1, 2])  # Measure all qubits

    # Simulate both circuits
    result_no_error = backend.run(qc_no_error, shots=shots_per_run).result()
    result_with_error = backend.run(qc_error, shots=shots_per_run).result()

    # Get measurement outcomes
    counts_no_error = result_no_error.get_counts()
    counts_with_error = result_with_error.get_counts()

    # Accumulate results for the "no error" circuit
    for key, value in counts_no_error.items():
        if key in cumulative_counts_no_error:
            cumulative_counts_no_error[key] += value
        else:
            cumulative_counts_no_error[key] = value

    # Accumulate results for the "with error" circuit
    for key, value in counts_with_error.items():
        if key in cumulative_counts_with_error:
            cumulative_counts_with_error[key] += value
        else:
            cumulative_counts_with_error[key] = value

# Normalize counts to probabilities
total_shots = shots_per_run * num_runs
normalized_counts_no_error = {key: value / total_shots for key, value in cumulative_counts_no_error.items()}
normalized_counts_with_error = {key: value / total_shots for key, value in cumulative_counts_with_error.items()}

# Plot the histograms to visualize the results
plot_histogram(
    [normalized_counts_no_error, normalized_counts_with_error],
    legend=['No Depolarizing Noise', 'With Depolarizing Noise'],
    title='Depolarizing Noise on CNOT Gates'
)
```
